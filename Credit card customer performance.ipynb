### Checking the credit card customer performance with the bank
#### A. Overview
Dataset 2 is a list bank customers who have a credit card with the bank. It suggests whether a credit card customer is performing good or bad for a bank. It contains 20 features including demographic, socio-economic and transactional variables for 5050 customers who have an active credit card with the bank.

Data description: The data contains a 'class' variable where bad=Customers who are not profitable credit card holders for the bank and Good=Customers who are profitable credit card holders for the bank. Customer_id is a unique serial number/identifier of each customer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('credit card customer data.csv')
df.head(5)
```

#### 2)  Checking the outliers using BoxPlot 

```python
def boxplot(cols,len1):
    plt.figure(1 , figsize = (25,5))
    n = 0
    for i in cols:
        n +=1
        plt.subplot(1,len1,n)
        plt.subplots_adjust(hspace=0.5, wspace=0.5)
        plt.title('Boxplot of {0}'.format(i))
        sns.boxplot(df[i])
    plt.show()
```

```python
length = 3
boxplot(['Customer_id','duration','credit_amount'],length)
boxplot(['installment_commitment','residence_since','age'],length)
boxplot(['bureau_score','num_dependents','Spend_debit_card'],length)
```

```python
def distplot(cols,len1):
    plt.figure(2 , figsize = (25,5))
    n = 0
    for i in cols:
        n +=1
        plt.subplot(1,len1,n)
        plt.subplots_adjust(hspace=0.5, wspace=0.5)
        plt.title('Distplot of {0}'.format(i))
        sns.distplot(df[i],bins=10)
    plt.show()
```

```python
length = 3
distplot(['Customer_id','duration','credit_amount'],length)
distplot(['installment_commitment','residence_since','age'],length)
distplot(['bureau_score','num_dependents','Spend_debit_card'],length)
```

#### Here we are distributing the customers by the age groups 

```python
def age_distribution():
    min_age = df['age'].min()
    age_18_25 = df.age[(df.age >=min_age) & (df.age<=25)]
    age_26_40 = df.age[(df.age >=26) & (df.age<=40)]
    age_41_55 = df.age[(df.age >=41) & (df.age<=55)]
    age_56_65 = df.age[(df.age >=56) & (df.age<=65)]
    age_66_above = df.age[(df.age >=66)]
    
    age_diff = ["18-25","26-40","41-55","56-65","65+"]
    total_Cust = [len(age_18_25),len(age_26_40),len(age_41_55),len(age_56_65),len(age_66_above)]

    return age_diff,total_Cust
```

```python
agex,agey = age_distribution()
plt.figure(figsize=(15,6))
sns.barplot(x=agex,y=agey)
plt.title(' Number of customer and ages ')
plt.xlabel('Age')
plt.ylabel('Number of Customer')
plt.show()
```

#### Develop a logistic regression or Random Forest model to estimate the relationship between 'class' and the other variables. Suggest variables which are more important.  

```python
df.info()
```

```python
df['other_payment_plans'] = df['other_payment_plans'].map({'none':'0','stores':'1','bank':'2'})
df['housing'] = df['housing'].map({'for free':'0','own':'1','rent':'2'})
df['own_telephone'] = df['own_telephone'].map({'none':'0','yes':'1'})
df['foreign_worker'] = df['foreign_worker'].map({'no':'0','yes':'1'})
df['class'] = df['class'].map({'bad':'0','good':'1'})
df.head(10)
```

```python
df['other_payment_plans'] = df['other_payment_plans'].astype(int)
df['housing'] = df['housing'].astype(int)
df['own_telephone'] = df['own_telephone'].astype(int)
df['foreign_worker'] = df['foreign_worker'].astype(int)
df['class'] = df['class'].astype(int)
```

```python
df.info()
```

```python
X = df[['duration', 'credit_amount', 'installment_commitment', 'residence_since', 'age', 'other_payment_plans', 'bureau_score', 'num_dependents', 'own_telephone', 'foreign_worker', 'Spend_debit_card']]

Y = df[['class']]
```

```python
X.head(5)
```

```python
Y.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1', 'l2', 'elasticnet'],'C':[0.01,2,3,4,6],'max_iter':[100,200,300]}

logistic_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=10)

logistic_regressor.fit(X_train,y_train)
```

```python
logistic_regressor.best_params_
```

```python
logistic_regressor.best_score_
```

```python
y_pred = logistic_regressor.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(classification_report(y_pred,y_test))
```

```python
cm = confusion_matrix(y_pred,y_test)

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
score = logistic_regressor.score(X_test,y_test)
score
```

### Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33,random_state=42)

treeclass = RandomForestClassifier(n_estimators=500,
                                  random_state = 42)
treeclass.fit(X_train, y_train)
```

```python
y_pred1 = treeclass.predict(X_test)
y_pred_prob = treeclass.predict_proba(X_test)
accuracy = treeclass.score(X_test, y_test)
accuracy
```

```python
cm = confusion_matrix(y_pred1,y_test)

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
score1 = accuracy_score(y_test,y_pred1)
score1
```

```python
print(classification_report(y_pred1,y_test))
```

### Comparison Logistic Regression Model and Random Forest model 

```python
x = ['Logistic Regresion Model Accuracy','Random Forest Model Accuracy']
y = [score,score1]
plt.figure(1,figsize=(10,8))
plt.bar(x,y,color ='maroon')
plt.title('Comparison Logistic Regression Model and Random Forest model',fontsize=14)
plt.xlabel('Model Name',fontsize=12)
plt.ylabel('Accuracy',fontsize=12)
plt.show()
```
